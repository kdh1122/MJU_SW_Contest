# 하이브 아티스트 크롤링

```python
#HIVE
import requests
from bs4 import BeautifulSoup
import pandas as pd

entertain = []
hive_artist = []

res = requests.get('https://hybecorp.com/kor/company/artist')
soup = BeautifulSoup(res.content, 'html.parser')
data = soup.select("body > section.page_wrap.about_wrap > div.company02.acticle_wrap.current > div.inner_con > ul.artist_list > li > a")

for item in data:
    name = item.get_text(strip=True)
    hive_artist.append(name)
    entertain.append("하이브")

print(hive_artist)
```

# JYP 아티스트 크롤링

```python
#JYP
import selenium
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

search_url = "https://www.jype.com/ko/Artist"

browser = webdriver.Chrome()  # chromedriver 다운받고, 다운 받은 경로 써주어야 함
browser.get(search_url)

browser.implicitly_wait(2)

# 요소를 가져올 CSS 선택자를 지정합니다.
css_selector = '#wrap > div > div.content_wrap.artist_list_wrap > div.artist_list_cont > ul > li.grid-item.on > div > strong'

target_elements = browser.find_elements(By.CSS_SELECTOR, css_selector)

# 각 요소의 텍스트를 리스트로 만듭니다.
jyp_artist = [element.text for element in target_elements]
for i in range(len(jyp_artist)):
    entertain.append("JYP Ent.")

# 텍스트 리스트를 출력합니다.
print(jyp_artist)

browser.quit()  # 작업이 끝났으면 브라우저를 닫습니다.
```

# SM 아티스트 크롤링

```python
#SM
search_url = "https://www.smentertainment.com/Entertainment/Artist"
search_url2 = "https://www.smentertainment.com/Entertainment/Artist?page=2"

browser = webdriver.Chrome()  # chromedriver 다운받고, 다운 받은 경로 써주어야 함
browser.get(search_url)

browser.implicitly_wait(2)

# 요소를 가져올 CSS 선택자를 지정합니다.
css_selector = '#contents > div.fullWrap > div > div.artist > div.artistList > ul > li > a > div.displayName > strong'

target_elements = browser.find_elements(By.CSS_SELECTOR, css_selector)

# 각 요소의 텍스트를 리스트로 만듭니다.
text_list = [element.text for element in target_elements]

browser.get(search_url2)

browser.implicitly_wait(2)

# 요소를 가져올 CSS 선택자를 지정합니다.
css_selector = '#contents > div.fullWrap > div > div.artist > div.artistList > ul > li > a > div.displayName > strong'

target_elements = browser.find_elements(By.CSS_SELECTOR, css_selector)

# 각 요소의 텍스트를 리스트로 만듭니다.
text_list2 = [element.text for element in target_elements]

# 텍스트 리스트를 출력합니다.
sm_artist = text_list+text_list2

for i in range(len(sm_artist)):
    entertain.append("에스엠")
print(sm_artist)


browser.quit()  # 작업이 끝났으면 브라우저를 닫습니다.
```
