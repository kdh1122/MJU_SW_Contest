```python
# 날짜 컬럼 값에 대해 년-월-일 이외의 시간 제거
sm[["날짜", "X"]] = sm["날짜"].str.split(" ", 1, expand=True)
sm.drop(columns=["X"], inplace=True)

hive[["날짜", "X"]] = hive["날짜"].str.split(" ", 1, expand=True)
hive.drop(columns=["X"], inplace=True)

fnc[["날짜", "X"]] = fnc["날짜"].str.split(" ", 1, expand=True)
fnc.drop(columns=["X"], inplace=True)

cube[["날짜", "X"]] = cube["날짜"].str.split(" ", 1, expand=True)
cube.drop(columns=["X"], inplace=True)

jyp[["날짜", "X"]] = jyp["날짜"].str.split(" ", 1, expand=True)
jyp.drop(columns=["X"], inplace=True)
```

```python
# 날짜 컬럼 형식 전처리
from datetime import datetime

def normalize_date(date_str):
    try:
        date_obj = datetime.strptime(date_str, "%Y.%m.%d.")
    except ValueError:
        try:
            date_obj = datetime.strptime(date_str, "%Y-%m-%d")
        except ValueError:
            return date_str

    normalized_date = date_obj.strftime("%Y-%m-%d")
    return normalized_date

sm["날짜"] = sm["날짜"].apply(normalize_date)
hive["날짜"] = hive["날짜"].apply(normalize_date)
fnc["날짜"] = fnc["날짜"].apply(normalize_date)
cube["날짜"] = cube["날짜"].apply(normalize_date)
jyp["날짜"] = jyp["날짜"].apply(normalize_date)
```

## 데이터 라벨링

```python
pos = []
neg = []

for word in pos_neg["긍정"]:
  pos.append(word)

for word in pos_neg["부정"]:
  neg.append(word)
```

```python
import re

# 각 단어에 따른 점수를 저장할 딕셔너리를 생성
word_scores = {word: 1 for word in pos}
word_scores.update({word: -1 for word in neg})

# "날짜" 컬럼을 기반으로 새로운 컬럼 "점수"를 추가
def calculate_score(title):
    score = 0
    # 영문자, 한글, 숫자를 제외한 문자를 모두 제거하여 단어만 남도록 처리
    words = re.findall(r"\b\w+\b", title)
    word_counts = {}  # 각 단어의 개수를 카운트하기 위한 딕셔너리 생성

    # 단어별 개수 카운트
    for i, word in enumerate(words):
        if word in word_scores:
            word_counts[word] = word_counts.get(word, 0) + 1

            # 단어의 앞뒤로 다른 단어도 함께 고려하여 점수 계산
            if i > 0:
                prev_word = words[i-1]
                combined_word = prev_word + word
                if combined_word in word_scores:
                    word_counts[combined_word] = word_counts.get(combined_word, 0) + 1
            if i < len(words) - 1:
                next_word = words[i+1]
                combined_word = word + next_word
                if combined_word in word_scores:
                    word_counts[combined_word] = word_counts.get(combined_word, 0) + 1

    # 각 단어별 개수에 따라 점수 더하기
    for word, count in word_counts.items():
        score += word_scores[word] * count
    return score

df["점수"] = df["제목"].apply(calculate_score)
df["라벨링"] = df["점수"].apply(lambda x: "1" if x > 0 else ("0" if x == 0 else "-1"))
```
