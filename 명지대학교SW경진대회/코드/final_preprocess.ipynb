```python
# @title 음악 차트 데이터
data_route1 = '/content/gdrive/MyDrive/ROF_gaisum_chart_1823.csv'
music = pd.read_csv(data_route1, encoding='utf-8') #euc-kr
music
```

```python
# @title 주식 데이터
data_route2 = '/content/gdrive/MyDrive/merged_stock.csv'
stock = pd.read_csv(data_route2, encoding='utf-8') #euc-kr
stock
```

```python
# @title 뉴스감정 데이터
data_route3 = '/content/gdrive/MyDrive/뉴스감정분석.csv'
news = pd.read_csv(data_route3, encoding='utf-8') #euc-kr
news
```

# 데이터 전처리

```python
# 날짜와 소속사 기준으로 Score 합
pre_music = music.groupby(['Date', 'entertain']).agg({'Score': ['sum', 'count']}).reset_index()

# 열 이름 변경
pre_music.columns = ['Date', 'entertain', 'ScoreSum', 'ScoreCount']
pre_music
```

```python
stock = stock.rename(columns={'날짜': 'Date'})
stock = stock.drop(columns = '전일비')
stock
```

```python
# 날짜 형식 통일
stock['Date'] = stock['Date'].str.replace('.', '-')
print(stock)
```

```python
# 주식 데이터와 음원차트 데이터 병합
merged_stock = stock.merge(pre_music, on=['Date', 'entertain'], how='left').fillna(0)
merged_stock['Date'] = pd.to_datetime(merged_stock['Date'])
merged_stock
```

# 주식+음원차트점수 데이터 csv저장

```python
# 주식+음원차트점수 데이터
merged_stock.to_csv('/content/gdrive/MyDrive/music_score_stock.csv', index=False)
```

# 뉴스 제목 감정 추가

```python
# 기사감정데이터 컬럼명 통합
news = news.drop(columns='제목')
news = news.rename(columns={'날짜': 'Date'})
news = news.rename(columns={'종목명': 'entertain'})
news
```

```python
# 날짜와 소속사 기준으로 기사 감정 합
news['Date'] = pd.to_datetime(news['Date'])
news = news.groupby(['Date', 'entertain'])['감정'].sum().reset_index()
news
```

```python
# 기사가 없는 날은 0으로 채워서 병합
fillna_stock = merged_stock.merge(news, on=['Date', 'entertain'], how='left').fillna(0)
fillna_stock
```

# 최종 데이터 저장

```python
# 최종 데이터
fillna_stock.to_csv('/content/gdrive/MyDrive/music_score_emotion_stock_fillna.csv', index=False)
```
