```python
#JUPYTER NOTEBOOK
from datetime import datetime, timedelta
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd

webdriver_path = '/path/to/chromedriver'
driver = webdriver.Chrome(executable_path=webdriver_path)

# 시작 날짜와 끝 날짜를 지정합니다.
start_date = datetime(2018, 10, 28)
end_date = datetime(2023, 12, 31)

# URL을 저장할 빈 리스트를 만듭니다.
urls = []

# 모든 날짜에 대한 URL을 생성합니다.
current_date = start_date
while current_date <= end_date:
    url = 'https://xn--o39an51b2re.com/melon/chart/daily/' + current_date.strftime('%Y%m%d')
    urls.append((current_date, url))
    current_date += timedelta(days=1)

# 크롤링한 결과를 저장할 빈 리스트를 만듭니다.
all_melon_chart_data = []

# 각 URL에 대해 크롤링을 수행합니다.
for date, url in urls:
    driver.get(url)
    driver.implicitly_wait(3)
    html = driver.page_source

    melon_chart_week = []
    soup = BeautifulSoup(html, 'html.parser')

    for i in range(100):
        ranks = soup.select_one(f'#chart-table-content > tbody > tr:nth-child({i+1}) > td.ranking > p:nth-child(1) > span').text.strip()
        titles = soup.select_one(f'#chart-table-content > tbody > tr:nth-child({i+1}) > td.subject > p:nth-child(1)').text.strip()
        artist = soup.select_one(f'#chart-table-content > tbody > tr:nth-child({i+1}) > td.subject > p.singer').text.strip()
        melon_chart_week.append([ranks, titles, artist])

    # 해당 날짜와 크롤링한 결과를 튜플로 묶어서 저장합니다.
    all_melon_chart_data.extend([(date, *data) for data in melon_chart_week])

# WebDriver를 종료해주는 것이 좋습니다.
driver.quit()

# 크롤링 결과를 DataFrame으로 만들어 CSV 파일로 저장합니다.
data = pd.DataFrame(all_melon_chart_data, columns=['Date', 'Rank', 'Title', 'Artist'])
data.to_csv('gaisum_chart_d_2018_2.csv', index=False)

print("finish")
data.head(100)

```
