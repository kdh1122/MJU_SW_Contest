## 에스엠 전처리

```python
# 에스엠 주식 데이터 전처리
import pandas as pd

sm = pd.read_csv("/content/drive/MyDrive/명지대학교SW경진대회/데이터/주식데이터/stock_sm.csv", encoding="UTF-8")

sm = sm.drop(labels="전일비", axis=1)
cd_sm = sm[sm["날짜"].str.contains("2023.07")].index
sm = sm.drop(cd_sm)
sm = sm.reset_index(drop=True)
sm["종목명"] = "에스엠"

columns = ["종가", "시가", "고가", "저가", "거래량"]
for column in columns:
    sm[column] = sm[column].str.replace(",", "").astype(int)

sm = sm[["날짜", "종목명", "종가", "시가", "고가", "저가", "거래량"]]
```

```python
sm.to_csv("sm.csv", encoding="utf-8-sig", index=False)
```
